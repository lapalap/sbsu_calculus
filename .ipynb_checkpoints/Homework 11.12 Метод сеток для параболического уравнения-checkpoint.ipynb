# Метод сеток для параболического уравнения

```python
import numpy as np
```

```python
n = 10
m = 100
tau = 0.001

r = 1
l = 0
h = (r - l) / n

def x_i(i):
    return l + i*h

def t_k(k):
    return k*tau

grid = np.array([[0 for i in range(n+1)] for j in range(m+1)], dtype = float)
grid_implicit = np.array([[0 for i in range(n+1)] for j in range(m+1)], dtype = float)
grid
```

```python
#Условие устойчивости
print(tau/(h*h)<=0.5)
```

```python
grid.shape
```

```python
def f(x,t):
    return -(x*(1-x))/((10+t)*(10+t)) + 2/(10+t)
```

Дано
$$f(x,t) = -\frac{x(1-x)}{(10+t)^2}+\frac{2}{10+t}$$
и условия
$$u(x,0) = \cos(\frac{x}{2})+ \frac{x(1-x)}{10}$$
$$u(0,t) = e^{-\frac{t}{4}}$$
$$u(1,t) = e^{-\frac{t}{4}} \cos{\frac{1}{2}}$$

```python
def psi_1(t):
    return np.exp(-t*0.25)

def psi_2(t):
    return np.exp(-t*0.25)*np.cos(0.5)

def phi(x):
    return np.cos(x*0.5)+x*(1-x)*0.1

def L_h(i,k):
    return (grid[k][i+1] - 2*grid[k][i]+grid[k][i-1])/(h*h)  
```

## Явная схема

```python
#1) находим u_i^0

for i in range(n+1):
    grid[0][i] = phi(x_i(i))
    
grid[0]
```

```python
grid
```

```python
#находим остальные слои

for j in range(1,m+1):
    for i in range(1,n):
        grid[j][i] = grid[j-1][i] +tau*((grid[j-1][i+1] - 2*grid[j-1][i]+grid[j-1][i-1])/(h*h)   + f(x_i(i),t_k(j-1)))

    grid[j][0] = psi_1(t_k(j))
    grid[j][n] = psi_2(t_k(j))

grid
```

## Неявная схема

```python
def TDMAsolver(a, b, c, d):
    nf = len(d)
    ac, bc, cc, dc = map(np.array, (a, b, c, d))
    for it in range(1, nf):
        mc = ac[it-1]/bc[it-1]
        bc[it] = bc[it] - mc*cc[it-1] 
        dc[it] = dc[it] - mc*dc[it-1]
        
    xc = bc
    xc[-1] = dc[-1]/bc[-1]

    for il in range(nf-2, -1, -1):
        xc[il] = (dc[il]-cc[il]*xc[il+1])/bc[il]

    return xc
```

```python
A = np.array([1./(h*h) for i in range(1,n)]+[0], dtype = float) 
B = np.array([1]+[-2./(h*h)-1./tau for i in range(1,n)]+[1], dtype = float)
C = np.array([0]+[1./(h*h) for i in range(1,n)], dtype = float)


for i in range(n+1):
    grid_implicit[0][i] = phi(x_i(i))

for j in range(1,m+1):
    D = np.array([psi_1(t_k(j))]+[-grid_implicit[j-1][i]/tau - f(x_i(i),t_k(j)) for i in range(1,n)]+[psi_2(t_k(j))], dtype = float)
    grid_implicit[j] = TDMAsolver(A, B, C, D)

grid_implicit
```

```python
A
```

```python
diff = abs(grid-grid_implicit)
diff
```

## Визуализация

```python
X = np.array([x_i(i)for j in range(0,m+1) for i in range(0, n+1)])
Y = np.array([t_k(k)for k in range(0,m+1) for j in range(0, n+1)])
Z = np.array([diff[k][i] for k in range(0,m+1) for i in range(0, n+1)])

Z
```
